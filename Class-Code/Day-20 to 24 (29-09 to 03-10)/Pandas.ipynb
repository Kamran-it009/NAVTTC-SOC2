```python
# Create dataframe
data = {'Company':['GOOGLE','GOOGLE','MICROSOFT','MICROSOFT','META','META'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
```

```python
df
```

```python
groups = df.groupby('Company')
```

```python
groups['Sales'].sum()
```

```python
group_ser = groups['Sales'].sum()
```

```python
group_ser.to_csv('Groups.csv')
```

```python
google = group_ser['GOOGLE']
meta = group_ser['META']
micro = group_ser['MICROSOFT']
```

```python
tables = [google, meta, micro]
names = ['google', 'meta', 'micro']
```

```python
record1 = {}
record2 = {}
record3 = {}
record1['google'] = tables[0]
record2['meta'] = tables[1]
record3['micro'] = tables[2]
```

```python
s1 = pd.Series(record1)
s2 = pd.Series(record2)
s3 = pd.Series(record3)
```

```python
s1.to_csv('google_record.csv')
s2.to_csv('meta_record.csv')
s3.to_csv('micro_record.csv')
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K2', 'K3', 'K4'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
left
```

```python
right
```

```python
# merging
pd.merge(left,right,how='inner',on='key')
```

```python
# merging
pd.merge(left,right,how='outer',on='key')
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right, how = 'inner')
```
