```python
def sum2(a = 10, b = 20): # Default parameters.
    x = a + b
    return x # returned value
```

```python
sum2()
```

```python
def sum3(a, b):
    print('a =', a)
    print('b =', b)
```

```python
sum3(34, 12) # positional arguments
```

```python
sum3(b=34, a=12) # keyword arguments
```

```python
sum2(10, 23)
```

```python
def sum(*a):
    s = 0
    for x in a:
        s = s + x
    return s
```

```python
sum(12)
```

```python
sum(12, 3,45)
```

```python
sum(13, 45, 67, 13, 45, 56, 100)
```

```python
def sum(**a):
    print(a)

sum(a = 5, b = 5, c = 6, d = 10, e = 67)
```

```python
# Map Functions
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)

print(list(squared_numbers))
```
